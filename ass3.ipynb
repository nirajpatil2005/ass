```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("sampleadd.csv")

# Convert 'Chance of Admit' to binary classification
df['Admit'] = (df['Chance of Admit'] >= 0.8).astype(int)

# 1. Scatter Plot: GRE Score vs CGPA
plt.figure(figsize=(6, 4))
sns.scatterplot(data=df, x="GRE Score", y="CGPA", hue="Admit", palette='Set1')
plt.title("Scatter Plot: GRE Score vs CGPA")
plt.tight_layout()
plt.show()

# 2. Bar Plot: Average CGPA by University Rating
plt.figure(figsize=(6, 4))
sns.barplot(x="University Rating", y="CGPA", data=df, estimator='mean', palette="coolwarm")
plt.title("Bar Plot: Average CGPA by University Rating")
plt.tight_layout()
plt.show()

# 3. Box Plot: CGPA across SOP scores
plt.figure(figsize=(6, 4))
sns.boxplot(x="SOP", y="CGPA", data=df, palette="Set2")
plt.title("Box Plot: CGPA by SOP")
plt.tight_layout()
plt.show()

# 4. Pie Chart: Distribution of Research (0 or 1)
plt.figure(figsize=(5, 5))
research_counts = df['Research'].value_counts()
plt.pie(research_counts, labels=research_counts.index, autopct='%1.1f%%', colors=["skyblue", "lightgreen"])
plt.title("Pie Chart: Research Experience")
plt.tight_layout()
plt.show()

# 5. Line Chart: TOEFL Score across students (simulated index)
plt.figure(figsize=(7, 4))
plt.plot(df.index, df["TOEFL Score"], marker='o', linestyle='--', color='purple')
plt.title("Line Chart: TOEFL Score by Applicant Index")
plt.xlabel("Applicant Index")
plt.ylabel("TOEFL Score")
plt.tight_layout()
plt.show()
```
