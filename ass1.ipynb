```python
import pandas as pd

# a) Read data from CSV and Excel (example)
df_csv = pd.read_csv("heart.csv")

# If you had an Excel file
# df_xls = pd.read_excel("heart.xlsx")

print("a) Data loaded successfully\n")

# b) Find shape of the data
print("b) Shape of data:", df_csv.shape)

# c) Find missing values
print("\nc) Missing values per column:\n", df_csv.isnull().sum())

# d) Find data type of each column
print("\nd) Data types:\n", df_csv.dtypes)

# e) Finding zeros in each column
print("\ne) Number of zeros in each column:")
zero_counts = (df_csv == 0).sum()
print(zero_counts)

# f) Indexing and selecting data
print("\nf) First 5 rows using indexing:")
print(df_csv.iloc[:5])  # or df_csv.head()

print("\nSorting data by 'Age':")
print(df_csv.sort_values(by='Age').head())

# g) Describe attributes of data
print("\ng) Summary statistics:")
print(df_csv.describe())

print("\nData types again:")
print(df_csv.dtypes)

# h) Counting unique values
print("\nh) Unique value counts for each column:")
for col in df_csv.columns:
    print(f"{col}: {df_csv[col].nunique()} unique values")

print("\nFormats of each column:")
for col in df_csv.columns:
    print(f"{col}: {type(df_csv[col].iloc[0])}")

# Example of converting data types (long <-> short)
print("\nConverting column 'Age' from int to string and back to int:")
df_csv['Age'] = df_csv['Age'].astype(str)
print("Type after conversion:", df_csv['Age'].dtype)

df_csv['Age'] = df_csv['Age'].astype(int)
print("Type after converting back:", df_csv['Age'].dtype)
```
